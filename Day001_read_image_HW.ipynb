思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 讀取影像
img_path = 'lena.png'
img_bgr = cv2.imread(img_path, cv2.IMREAD_COLOR);
# seperate BGR channel
img_b,img_g,img_r = np.zeros((512,512,3),dtype=np.uint8),np.zeros((512,512,3),dtype=np.uint8),np.zeros((512,512,3),dtype=np.uint8)
img_b[...,0],img_g[...,1],img_r[...,2] = img_bgr[..., 0], img_bgr[..., 1], img_bgr[..., 2]

# print(img_b.shape)--->(512,3)
plt.figure(figsize=(20,40))
plt.subplot(10,5,1)
plt.imshow(img_b)
plt.axis('on')
plt.title("img_b")

plt.subplot(10,5,2)
plt.imshow(img_g)
plt.axis('on')
plt.title("img_g")

plt.subplot(10,5,3)
plt.imshow(img_r)
plt.axis('on')
plt.title("img_r")
```
